#Analise da categoria de Vinho de Mesa

```python
import matplotlib.ticker as ticker

Tinto_mesa = dados_vinho.T['Tinto_mesa']
Rosado_mesa = dados_vinho.T['Rosado_mesa']
Branco_mesa = dados_vinho.T['Branco_mesa']

# Crie um gráfico com as três colunas selecionadas.
plt.figure(figsize=(12, 6))
plt.plot(Tinto_mesa, label='Tinto_mesa')
plt.plot(Rosado_mesa, label='Rosado_mesa')
plt.plot(Branco_mesa, label='Branco_mesa')

# Adicione legendas, rótulos e outras configurações do gráfico, se necessário.
plt.legend()
plt.title('Vinho de Mesa')
plt.xlabel('Ano')
plt.ylabel('Total de Volume Comercializado')
plt.grid(linestyle='--')
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
plt.show()
```

#PIB x Total comercializado

```python
#pib = pd.read_csv("/content/pib.csv", sep = ';')
```

```python
populacao = pd.read_excel("/content/populacao.xlsx")
```

```python
populacao
```

```python
populacao = populacao.set_index("Ano")
populacao.head()
```

```python
total_versus_pop = totais.join(populacao)
```

```python
total_versus_pop.head()
```

```python
total_versus_pop.columns
```

```python
total_versus_pop['Litros_per_capita'] = (total_versus_pop['total_comercializado']) / total_versus_pop['Rio Grande do Sul']
```

```python
total_versus_pop.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10,6))
paleta = sns.color_palette("crest", 14 )
sns.barplot(data=total_versus_pop, x= total_versus_pop.index, y='Litros_per_capita',palette = paleta)
plt.grid( True, axis='y',linestyle='--', alpha=0.7)
plt.title("Quantidade de litros per capita")

plt.ylim(0,30)
plt.show()

```

criar uma coluna com a porcentagem do valor comercializado corresponde do pib anual.

porcentagem da comercialização em relação ao pib  = valor de comercialização do ano * 100 / valor do pib do ano
